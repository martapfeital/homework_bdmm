## Type and Missing Values Analysis

```python
def check_all_field_types_with_nulls(db, collection_name, fields):
    collection = db[collection_name]
    total_docs = collection.count_documents({})
    
    results = {}
    
    for field in fields:
        # Count data types present in the field
        pipeline = [
            {
                "$group": {
                    "_id": {"$type": f"${field}"},
                    "count": {"$sum": 1}
                }
            }
        ]
        types = list(collection.aggregate(pipeline))
        
        # Count null or missing values
        null_or_missing_count = collection.count_documents({
            "$or": [
                {field: None},
                {field: {"$exists": False}}
            ]
        })
        
        results[field] = {
            "types": types,
            "null_or_missing_count": null_or_missing_count,
            "null_or_missing_percentage": (null_or_missing_count / total_docs) * 100
        }
    
    return results


collection_name = "listingsAndReviews_HW2"

types_info = check_all_field_types_with_nulls(db, collection_name, fields)

for field, info in types_info.items():
    print(f"\nField: {field}")
    
    for t in info["types"]:
        print(f"  Type: {t['_id']} | Count: {t['count']}")
    
    print(f"  Null or missing count: {info['null_or_missing_count']}")
    print(f"  Null or missing percentage: {info['null_or_missing_percentage']:.2f}%")
```

##  Duplicate Analysis

```python
def find_duplicates(db, collection_name, fields):
    collection = db[collection_name]

    # Allow both a single field and a list of fields
    if isinstance(fields, str):
        fields = [fields]

    # Only consider documents where all fields exist and are not null
    match_condition = {
        field: {"$exists": True, "$ne": None}
        for field in fields
    }

    # Build composite key for grouping
    group_id = {
        field.replace(".", "_"): f"${field}"
        for field in fields
    }

    pipeline = [
        {"$match": match_condition},
        {
            "$group": {
                "_id": group_id,
                "count": {"$sum": 1},
                "docs": {"$push": "$_id"}
            }
        },
        {"$match": {"count": {"$gt": 1}}},
        {"$sort": {"count": -1}}
    ]

    return list(collection.aggregate(pipeline))


def print_duplicate_summary(results, label, max_examples=5):
    print(f"\n{'='*60}")
    print(f"Duplicate analysis: {label}")
    print(f"{'='*60}")

    if not results:
        print("No duplicates found.")
        return

    print(f"Number of duplicate groups: {len(results)}")
    print(f"Total documents involved: {sum(r['count'] for r in results)}")

    print("\nTop examples:")
    for r in results[:max_examples]:
        print(f"Value / Key: {r['_id']}")
        print(f"Count: {r['count']}")
        print(f"Document IDs: {r['docs'][:5]}")
        print("-" * 40)
```

```python
collection_name = "listingsAndReviews_HW2"

duplicates_listing_url = find_duplicates(db, collection_name, "listing_url")
print_duplicate_summary(duplicates_listing_url, "listing_url")

duplicates_name_host = find_duplicates(
    db,
    collection_name,
    ["name", "host.host_id"]
)
print_duplicate_summary(duplicates_name_host, "name + host.host_id")

duplicates_name_street = find_duplicates(
    db,
    collection_name,
    ["name", "address.street"]
)
print_duplicate_summary(duplicates_name_street, "name + address.street")

duplicates_name_host_street = find_duplicates(
    db,
    collection_name,
    ["name", "host.host_id", "address.street"]
)
print_duplicate_summary(duplicates_name_host_street, "name + host.host_id + address.street")
```

## Reviews Redesign

```python
import datetime

collection = db["listingsAndReviews_HW2"]

reviews_to_insert = []

for listing in collection.find({}, {"reviews": 1}):
    listing_id = listing["_id"]
    reviews = listing.get("reviews", [])

    if not isinstance(reviews, list) or len(reviews) == 0:
        continue

    # ordenar por data (mais recentes primeiro)
    reviews_sorted = sorted(
        reviews,
        key=lambda r: r.get("date", datetime.datetime.min),
        reverse=True
    )

    # subset (top 3 mais recentes)
    reviews_subset = [
        {
            "date": r.get("date"),
            "reviewer_name": r.get("reviewer_name"),
            "comments": r.get("comments")
        }
        for r in reviews_sorted[:3]
    ]

    # preparar todas as reviews para nova coleção
    for r in reviews:
        review_doc = r.copy()
        review_doc["listing_id"] = listing_id  # garantir consistência
        reviews_to_insert.append(review_doc)

    # atualizar listing
    collection.update_one(
        {"_id": listing_id},
        {
            "$set": {"reviews_subset": reviews_subset},
            "$unset": {"reviews": ""}
        }
    )

# inserir na nova coleção
if reviews_to_insert:
    db.reviews.insert_many(reviews_to_insert)
    print("Inserted reviews:", len(reviews_to_insert))
else:
    print("No reviews to insert")
```

# Exercise B2: Review Quality and Activity Score

## B2 – Assumptions & Decisions

1. **City**: New York only (address.market == "New York")
2. **Minimum reviews**: Properties with fewer than 10 reviews are excluded
3. **Rating field**: review_scores.review_scores_rating (sub-field of embedded
   review_scores document, scale 0–10), normalized to 0–1 by dividing by 10
4. **last_review** on the listing document is used for recency (pre-computed,
   already indexed) rather than querying the reviews collection
5. **Missing ratings**: Properties where review_scores_rating is null or missing
   are excluded entirely since rating is the most heavily weighted component
6. **Recency score**: 1 = property whose last_review equals the global maximum
   date across all qualifying NY properties; 0 = property furthest from that date.
   Formula: 1 - (days_since / max_days_since)
7. **Normalized review count**: computed within each property_type group,
   so a property is compared only against others of the same type
8. **Weekly recalibration**: a compound index on (address.market,
   number_of_reviews, last_review, property_type) is created to support
   repeated efficient execution of this query
9. **Property types with fewer than 3 listings** after filtering are excluded
   from box plots (insufficient data for a meaningful distribution) but kept
   in the summary table

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import warnings, time
from pymongo import MongoClient, ASCENDING, DESCENDING
warnings.filterwarnings("ignore")

client   = MongoClient("mongodb://AzureDiamond:hunter2@localhost:27017")
db       = client.sample_airbnb
listings = db["listingsAndReviews_HW2"]

# ════════════════════════════════════════════════════════════════
# STEP 1 – RAW PIPELINE (before optimization)
# ════════════════════════════════════════════════════════════════

pipeline_b2 = [
    # Stage 1: filter New York, min 10 reviews, rating present, last_review present
    {
        "$match": {
            "address.market": "New York",
            "number_of_reviews": {"$gte": 10},
            "review_scores.review_scores_rating": {"$exists": True, "$ne": None},
            "last_review": {"$exists": True, "$ne": None},
        }
    },
    # Stage 2: project only needed fields
    {
        "$project": {
            "property_type":     1,
            "number_of_reviews": 1,
            "last_review":       1,
            "rating": "$review_scores.review_scores_rating",
        }
    },
]

# ── Run raw pipeline and time it ─────────────────────────────
t0 = time.time()
raw_docs = list(listings.aggregate(pipeline_b2))
t1 = time.time()
raw_time_ms = (t1 - t0) * 1000
print(f"Raw pipeline: {len(raw_docs)} docs in {raw_time_ms:.1f} ms")

# ── Explain raw pipeline ──────────────────────────────────────
explain_raw = db.command(
    "explain",
    {
        "aggregate": "listingsAndReviews_HW2",
        "pipeline":  pipeline_b2,
        "cursor":    {}
    },
    verbosity="executionStats"
)

#raw_cursor     = explain_raw["stages"][0]["$cursor"]
raw_exec       = explain_raw["executionStats"]
raw_plan       = explain_raw["queryPlanner"]["winningPlan"]

# Walk down inputStage chain to find the scan stage
def get_leaf_stage(plan):
    while "inputStage" in plan:
        plan = plan["inputStage"]
    return plan

raw_leaf = get_leaf_stage(raw_plan)

raw_metrics = {
    "Version":        "Before Optimization",
    "Exec Time (ms)": raw_exec.get("executionTimeMillis", "N/A"),
    "Docs Examined":  raw_exec.get("totalDocsExamined",   "N/A"),
    "Keys Examined":  raw_exec.get("totalKeysExamined",   "N/A"),
    "Docs Returned":  raw_exec.get("nReturned",           "N/A"),
    "Stage Type":     raw_leaf.get("stage",               "N/A"),
    "Index Used":     raw_leaf.get("indexName",           "none"),
}
print("Raw metrics:", raw_metrics)
```

```python
# ════════════════════════════════════════════════════════════════
# STEP 2 – CREATE OPTIMIZED INDEX
# Compound index covers the $match filter fields in order:
#   1. address.market  (equality)
#   2. number_of_reviews (range: >= 10)
#   3. last_review  (used in projection + recency calc)
#   4. property_type (used in grouping)
# This allows MongoDB to satisfy both $match and $project
# from the index, minimizing document fetches.
# ════════════════════════════════════════════════════════════════

listings.create_index(
    [
        ("address.market",                       ASCENDING),
        ("number_of_reviews",                    ASCENDING),
        ("last_review",                          DESCENDING),
        ("property_type",                        ASCENDING),
        ("review_scores.review_scores_rating",   ASCENDING),
    ],
    name="idx_b2_ny_score"
)
print("B2 index created: idx_b2_ny_score")

# ── Run optimized pipeline and time it ───────────────────────
t0 = time.time()
opt_docs = list(listings.aggregate(pipeline_b2))
t1 = time.time()
opt_time_ms = (t1 - t0) * 1000
print(f"Optimized pipeline: {len(opt_docs)} docs in {opt_time_ms:.1f} ms")

# ── Explain optimized pipeline ────────────────────────────────
explain_opt = db.command(
    "explain",
    {
        "aggregate": "listingsAndReviews_HW2",
        "pipeline":  pipeline_b2,
        "cursor":    {}
    },
    verbosity="executionStats"
)

#opt_cursor = explain_opt["stages"][0]["$cursor"]
opt_exec   = explain_opt["executionStats"]
opt_plan   = explain_opt["queryPlanner"]["winningPlan"]
opt_leaf   = get_leaf_stage(opt_plan)

opt_metrics = {
    "Version":        "After Optimization",
    "Exec Time (ms)": opt_exec.get("executionTimeMillis", "N/A"),
    "Docs Examined":  opt_exec.get("totalDocsExamined",   "N/A"),
    "Keys Examined":  opt_exec.get("totalKeysExamined",   "N/A"),
    "Docs Returned":  opt_exec.get("nReturned",           "N/A"),
    "Stage Type":     opt_leaf.get("stage",               "N/A"),
    "Index Used":     opt_leaf.get("indexName",           "none"),
}
print("Opt metrics:", opt_metrics)
```

```python
# ════════════════════════════════════════════════════════════════
# STEP 3 – COMPUTE SCORE IN PYTHON
# (cross-property values like max_per_type and global_max_date
#  require two passes; cleaner and more flexible in Python)
# ════════════════════════════════════════════════════════════════

def safe_float(v):
    try:
        return float(str(v))
    except Exception:
        return np.nan

df = pd.DataFrame(opt_docs)
df["rating"]       = df["rating"].apply(safe_float)
df["review_count"] = df["number_of_reviews"].apply(safe_float)
df["last_review"]  = pd.to_datetime(df["last_review"], errors="coerce")
df = df.dropna(subset=["rating", "last_review", "property_type"])
print(f"Records after cleaning: {len(df)}")

# Global latest review date across all qualifying NY properties
global_max_date = df["last_review"].max()
print(f"Global latest review date: {global_max_date.date()}")

# Days since latest review (0 = most recent)
df["days_since"] = (global_max_date - df["last_review"]).dt.days
max_days = df["days_since"].max()

# Recency score: 1 for most recent, 0 for oldest
df["recency_score"] = 1 - (df["days_since"] / max_days)

# Normalized review count within each property_type group
df["max_count_in_type"] = df.groupby("property_type")["review_count"].transform("max")
df["norm_review_count"] = df["review_count"] / df["max_count_in_type"]

# Rating normalized to 0–1 (scale is 0–100)
df["rating_norm"] = df["rating"] / 100.0

# Final score
df["score"] = (
    df["rating_norm"]       * 0.5 +
    df["norm_review_count"] * 0.3 +
    df["recency_score"]     * 0.2
)

print("\nScore distribution:")
df["score"].describe().round(4)
```

```python
# ════════════════════════════════════════════════════════════════
# STEP 4 – SUMMARY TABLE per property type
# ════════════════════════════════════════════════════════════════

summary = (
    df.groupby("property_type")
    .agg(
        Count         = ("score", "count"),
        Mean_Score    = ("score", lambda x: round(x.mean(),   4)),
        Median_Score  = ("score", lambda x: round(x.median(), 4)),
        Min_Score     = ("score", lambda x: round(x.min(),    4)),
        Max_Score     = ("score", lambda x: round(x.max(),    4)),
        Std_Score     = ("score", lambda x: round(x.std(),    4)),
    )
    .sort_values("Count", ascending=False)
    .reset_index()
)

print("\nB2 – Review Quality & Activity Score per Property Type (New York):")
print(summary.to_string(index=False))

# ── Render summary as figure table ───────────────────────────
fig_tbl, ax_tbl = plt.subplots(figsize=(13, len(summary) * 0.45 + 1))
ax_tbl.axis("off")

tbl = ax_tbl.table(
    cellText  = summary.values,
    colLabels = summary.columns,
    cellLoc   = "center",
    loc       = "center",
)
tbl.auto_set_font_size(False)
tbl.set_fontsize(9)
tbl.scale(1, 1.5)

for (row, col), cell in tbl.get_celld().items():
    if row == 0:
        cell.set_facecolor("#2A9D8F")
        cell.set_text_props(color="white", fontweight="bold")
    elif row % 2 == 0:
        cell.set_facecolor("#f0f0f0")
    else:
        cell.set_facecolor("white")

fig_tbl.suptitle("B2 – Score Summary per Property Type – New York",
                 fontsize=11, fontweight="bold")
plt.tight_layout()
plt.show()
```

# Exercise B3: Long-Term Rental Suggestion

## B3 – Assumptions & Decisions

1. **Review as occupancy proxy**: A review in a given month is treated as
   evidence the property was occupied that month. No review = potential vacancy.

2. **Vacancy definition**: A month is considered vacant for a property in a
   given year if zero reviews were written that month, AND the property had
   at least one review in that year (i.e. the property was active that year).
   Properties with no reviews at all in a year are excluded from that year's
   analysis to avoid counting inactive/delisted properties as vacant.

3. **Recurring vacancy**: A month is flagged as a recurring vacancy if the
   property was vacant in that month in at least 50% of its active years.
   This threshold balances robustness with inclusivity.

4. **Consecutive month requirement**: Only runs of 2 or more consecutive
   recurring vacant months qualify for a rental_suggestion. A single isolated
   vacant month is not enough — the task requires "more than one consecutive
   month without reviews".

5. **Minimum years**: A property must have review data spanning at least 2
   distinct years to establish a recurring pattern. Single-year properties
   cannot demonstrate recurrence and are excluded.

6. **Data source**: The reviews collection (separated during redesign) is used,
   with each document containing listing_id and date fields.

7. **Cities**: Hong Kong, Montreal, Barcelona — matched via address.market
   on the listings collection.

8. **Database design**: rental_suggestion is stored as an array of month
   numbers (1–12) on each listing document in listingsAndReviews_HW2.
   It is intended to be refreshed once per year via a bulk update.
   An index on address.market already exists to support city-filtered queries.

9. **Scatter plot**: Points with count = 0 (no suggestions for that city/month)
   are omitted from the plot to avoid clutter. Point size is scaled
   proportionally to the number of properties suggested for that month.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from collections import defaultdict
from pymongo import MongoClient, ASCENDING, UpdateOne
warnings.filterwarnings("ignore")

client   = MongoClient("mongodb://AzureDiamond:hunter2@localhost:27017")
db       = client.sample_airbnb
listings = db["listingsAndReviews_HW2"]
reviews  = db["reviews"]

B3_CITIES = ["Hong Kong", "Montreal", "Barcelona"]

MONTH_NAMES = ["Jan","Feb","Mar","Apr","May","Jun",
               "Jul","Aug","Sep","Oct","Nov","Dec"]

CITY_COLORS = {
    "Hong Kong": "#E63946",
    "Montreal":  "#2A9D8F",
    "Barcelona": "#E9C46A",
}

```

```python
# ════════════════════════════════════════════════════════════════
# STEP 1 – Get listings in the 3 cities
# ════════════════════════════════════════════════════════════════

city_listings = list(listings.aggregate([
    {"$match": {"address.market": {"$in": B3_CITIES}}},
    {"$project": {"city": "$address.market"}}
]))

listing_id_to_city = {doc["_id"]: doc["city"] for doc in city_listings}
all_listing_ids    = list(listing_id_to_city.keys())
print(f"Listings in 3 cities: {len(all_listing_ids)}")
```

```python
# ════════════════════════════════════════════════════════════════
# STEP 2 – Fetch review dates from reviews collection
# ════════════════════════════════════════════════════════════════

rev_docs = list(reviews.aggregate([
    {
        "$match": {
            "listing_id": {"$in": all_listing_ids},
            "date":        {"$exists": True, "$ne": None},
        }
    },
    {
        "$project": {
            "listing_id": 1,
            "date":       1,
            "_id":        0,
        }
    },
]))
print(f"Review documents fetched: {len(rev_docs)}")

# Build dict: listing_id → list of (year, month)
reviews_by_listing = defaultdict(list)
for doc in rev_docs:
    dt = doc.get("date")
    if dt:
        try:
            reviews_by_listing[doc["listing_id"]].append((dt.year, dt.month))
        except AttributeError:
            pass

```

```python
# ════════════════════════════════════════════════════════════════
# STEP 3 – Core vacancy logic
# ════════════════════════════════════════════════════════════════

def find_consecutive_groups(months, min_length=2):
    """
    Given a sorted list of month numbers, return only those months
    that belong to a consecutive run of at least min_length months.
    Handles year wrap-around: month 12 and month 1 are consecutive.
    """
    if not months:
        return []

    months = sorted(set(months))

    # Check for year wrap-around (e.g. Nov, Dec, Jan, Feb)
    # by treating the list as circular
    # Build groups considering wrap
    extended = months + [m + 12 for m in months]  # extend for circularity
    groups   = []
    group    = [extended[0]]

    for m in extended[1:]:
        if m == group[-1] + 1:
            group.append(m)
        else:
            groups.append(group)
            group = [m]
    groups.append(group)

    # Collect months from long-enough groups, mapped back to 1–12
    result = set()
    for g in groups:
        if len(g) >= min_length:
            for m in g:
                result.add(((m - 1) % 12) + 1)

    # Only return months that were in the original list
    return sorted(result & set(months))


def get_rental_suggestion(yr_month_list,
                           min_years=2,
                           vacancy_threshold=0.5,
                           consec_threshold=2):
    """
    Returns list of month numbers (1–12) suitable for long-term rental.
    """
    if not yr_month_list:
        return []

    # Group months by year
    by_year = defaultdict(set)
    for (yr, mo) in yr_month_list:
        by_year[yr].add(mo)

    years = sorted(by_year.keys())

    if len(years) < min_years:
        return []

    # For each year the property was active, find vacant months
    vacancy_count = defaultdict(int)
    active_years  = 0

    for yr in years:
        months_with_reviews = by_year[yr]
        # Only count this year if the property had at least 1 review
        if not months_with_reviews:
            continue
        active_years += 1
        for mo in range(1, 13):
            if mo not in months_with_reviews:
                vacancy_count[mo] += 1

    if active_years < min_years:
        return []

    # Recurring vacant = vacant in >= 50% of active years
    recurring_vacant = sorted([
        mo for mo, cnt in vacancy_count.items()
        if cnt / active_years >= vacancy_threshold
    ])

    # Keep only months in consecutive runs of >= 2
    return find_consecutive_groups(recurring_vacant, min_length=consec_threshold)


# Compute suggestion for every listing
suggestion_map = {}
for lid in all_listing_ids:
    yr_mo = reviews_by_listing.get(lid, [])
    suggestion_map[lid] = get_rental_suggestion(yr_mo)

n_with = sum(1 for v in suggestion_map.values() if v)
print(f"Listings with a rental suggestion: {n_with} / {len(all_listing_ids)}")

```

```python
# ════════════════════════════════════════════════════════════════
# STEP 4 – Write rental_suggestion to MongoDB
# ════════════════════════════════════════════════════════════════

bulk_ops = [
    UpdateOne(
        {"_id": lid},
        {"$set": {"rental_suggestion": months}}
    )
    for lid, months in suggestion_map.items()
]

if bulk_ops:
    result = listings.bulk_write(bulk_ops)
    print(f"Documents updated: {result.modified_count}")

# Quick verification
sample = listings.find_one({
    "address.market":   {"$in": B3_CITIES},
    "rental_suggestion": {"$not": {"$size": 0}},
})
if sample:
    print(f"Sample – {sample['_id']}: months {sample['rental_suggestion']}")

```

```python
# ════════════════════════════════════════════════════════════════
# STEP 5 – Aggregate counts per city × month for scatter plot
# ════════════════════════════════════════════════════════════════

city_month_counts = defaultdict(lambda: defaultdict(int))

for lid, months in suggestion_map.items():
    city = listing_id_to_city.get(lid)
    if city in B3_CITIES:
        for mo in months:
            city_month_counts[city][mo] += 1

# Build DataFrame (all 12 months × 3 cities)
rows = []
for city in B3_CITIES:
    for mo in range(1, 13):
        cnt = city_month_counts[city].get(mo, 0)
        rows.append({"city": city, "month": mo, "count": cnt})

df_b3 = pd.DataFrame(rows)

print("\nNon-zero suggestions:")
print(df_b3[df_b3["count"] > 0].to_string(index=False))

```

```python
# ════════════════════════════════════════════════════════════════
# STEP 6 – Scatter Plot
# ════════════════════════════════════════════════════════════════

fig, ax = plt.subplots(figsize=(14, 6))

for city in B3_CITIES:
    sub      = df_b3[(df_b3["city"] == city) & (df_b3["count"] > 0)]
    sizes    = sub["count"] * 30   # scale bubble size

    sc = ax.scatter(
        sub["month"],
        sub["count"],
        s         = sizes,
        color     = CITY_COLORS[city],
        alpha     = 0.75,
        edgecolors= "white",
        linewidths= 0.8,
        label     = city,
        zorder    = 3,
    )

    # Annotate count above each point
    for _, row in sub.iterrows():
        ax.annotate(
            int(row["count"]),
            (row["month"], row["count"]),
            textcoords = "offset points",
            xytext     = (0, 8),
            ha         = "center",
            fontsize   = 8,
            fontweight = "bold",
            color      = CITY_COLORS[city],
        )

ax.set_xticks(range(1, 13))
ax.set_xticklabels(MONTH_NAMES, fontsize=10)
ax.set_xlabel("Month of the Year", fontsize=12)
ax.set_ylabel("Number of Properties with Long-Term Rental Suggestion", fontsize=11)
ax.set_title(
    "Long-Term Rental Suggestions by Month and City\n"
    "Hong Kong  ·  Montreal  ·  Barcelona\n"
    "(point size proportional to number of properties)",
    fontsize=13, fontweight="bold"
)
ax.legend(fontsize=11, frameon=True, loc="upper right")
ax.grid(axis="both", linestyle="--", alpha=0.35)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_xlim(0.5, 12.5)

# Add a size legend
for n, label in [(5, "5 props"), (20, "20 props"), (50, "50 props")]:
    ax.scatter([], [], s=n*30, color="gray", alpha=0.5, label=label)
ax.legend(fontsize=10, frameon=True, loc="upper right",
          title="City / Count", title_fontsize=9)

plt.tight_layout()
plt.show()
```
